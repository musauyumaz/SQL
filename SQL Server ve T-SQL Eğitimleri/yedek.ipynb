# [dbo].[VW_GOTUR]

```sql
/****** Object:  View [dbo].[VW_GOTUR]    Script Date: 25.06.2023 16:28:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW  [dbo].[VW_GOTUR]
AS 
SELECT P.Adi + ' ' + P.SoyAdi [ADI SOYADI], K.KategoriAdi [KATEGORİ ADI], COUNT(S.SatisID) [TOPLAM SATIŞ] FROM Personeller P 
INNER JOIN Satislar S ON P.PersonelID = S.PersonelID 
INNER JOIN [Satis Detaylari] SD ON S.SatisID = SD.SatisID 
INNER JOIN Urunler U ON U.UrunID = SD.UrunID 
INNER JOIN Kategoriler K ON K.KategoriID = U.KategoriID 
GROUP BY P.Adi + ' ' + P.SoyAdi , K.KategoriAdi
GO
```

# [dbo].[Faturalar]

```sql
/****** Object:  View [dbo].[Faturalar]    Script Date: 25.06.2023 16:28:30 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

create view [dbo].[Faturalar] AS
SELECT Satislar.SevkAdi, Satislar.SevkAdresi, Satislar.SevkSehri, Satislar.SevkBolgesi, Satislar.SevkPostaKodu, 
	Satislar.SevkUlkesi, Satislar.MusteriID, Musteriler.SirketAdi AS MusteriName, Musteriler.Adres, Musteriler.Sehir, 
	Musteriler.Bolge, Musteriler.PostaKodu, Musteriler.Ulke, 
	(Adi + ' ' + SoyAdi) AS PersonelSatislari, 
	Satislar.SatisID, Satislar.SatisTarihi, Satislar.OdemeTarihi, Satislar.SevkTarihi, Nakliyeciler.SirketAdi As NakliyeciName, 
	"Satis Detaylari".UrunID, Urunler.UrunAdi, "Satis Detaylari".BirimFiyati, "Satis Detaylari".Miktar, 
	"Satis Detaylari".İndirim, 
	(CONVERT(money,("Satis Detaylari".BirimFiyati*Miktar*(1-İndirim)/100))*100) AS ExtendedPrice, Satislar.NakliyeUcreti
FROM 	Nakliyeciler INNER JOIN 
		(Urunler INNER JOIN 
			(
				(Personeller INNER JOIN 
					(Musteriler INNER JOIN Satislar ON Musteriler.MusteriID = Satislar.MusteriID) 
				ON Personeller.PersonelID = Satislar.PersonelID) 
			INNER JOIN "Satis Detaylari" ON Satislar.SatisID = "Satis Detaylari".SatisID) 
		ON Urunler.UrunID = "Satis Detaylari".UrunID) 
	ON Nakliyeciler.NakliyeciID = Satislar.ShipVia
GO
```
